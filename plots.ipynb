```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xls = pd.ExcelFile('/mnt/j/OneDrive/1.Studia/1.Studia_4/Publication/scRNA_benchmarks.xlsx')
df1 = pd.read_excel(xls, 'Sheet2', header=None, index_col=None)
df2 = df1.to_numpy()
```

## Silhouette scores

```python
kmean_silh_temp = df2[:, 0]
kmean_silh = kmean_silh_temp[~np.isnan(kmean_silh_temp)]
qa_silh_temp = df2[:, 1]
qa_silh = qa_silh_temp[~np.isnan(qa_silh_temp)]

del kmean_silh_temp, qa_silh_temp
```

```python
x = kmean_silh
factor = kmean_silh.size/qa_silh.size
y = np.repeat(qa_silh, int(factor))

print("kmeans data size: ", x.size, ", qa data size: ", qa_silh.size)

bins = np.linspace(0, 0.04, 50)
fig, ax = plt.subplots(figsize=(16, 6))

ax.hist(x, bins, alpha=0.8, label='kmeans')

ymin, ymax = ax.get_ylim()
ax.set_ylim(ymin, ymax)

ax.vlines(x = qa_silh, ymin = ymin, ymax = ymax,
           colors = 'purple',
           label = 'QA')

plt.xlabel('Silhouette Score', fontsize=18)
plt.ylabel('Frequency of K-means scores', fontsize=18)
plt.title('Comparison of K-means with Quantum Annealing-based Clustering ', fontsize=22)
plt.legend(loc='upper right', framealpha=1, frameon=True, shadow=True, borderpad=1, fontsize=14)
plt.show()
```

# Number of Kmeans runs to achieve QA score

```python
kmean_n_iter_temp = df2[:, 6]
kmean_n_iter = kmean_n_iter_temp[~np.isnan(kmean_n_iter_temp)]
kmean_n_iter.shape[0]
```

```python
bins = kmean_n_iter.shape[0]

fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(kmean_n_iter, bins, alpha=0.8, label='kmeans',rwidth=0.85);

plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Number of repetitions', fontsize=14)
plt.title('Number of K-means repetitions to achive QA accuracy', fontsize=18)
# plt.legend(loc='upper right', framealpha=1, frameon=True, shadow=True, borderpad=1, fontsize=14)
plt.text(120, 6,'Number of trails: 30 \nMean value: 48.52', fontsize = 18, 
         bbox = dict(facecolor = 'grey', alpha = 0.5))
plt.show()
```
